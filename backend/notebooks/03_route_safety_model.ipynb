## Tujuan
Membangun model klasifikasi yang memprediksi tingkat keamanan rute wisata  
di Dieng berdasarkan kondisi fisik jalan, cuaca, dan jenis kendaraan.

**Dataset:** Custom Hyper-Local (riset lapangan tim)
- 15 rute wisata di kawasan Dieng
- 3 jenis kendaraan (motor, mobil, bus)
- 4 kondisi cuaca (cerah, mendung, hujan, kabut)
- Total: 180 sampel (data augmentation)

```python
import sys
sys.path.insert(0, '..')

from app.models.train_route_model import generate_route_dataset, train_route_safety_model
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print('Modules loaded!')
```

## 1. Generate & Explore Dataset

```python
df = generate_route_dataset()
print(f'\nDataset shape: {df.shape}')
print(f'\nColumns: {list(df.columns)}')
df.head(10)
```

```python
# Visualize safety distribution
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Safety class distribution
safety_labels = {0: 'Aman', 1: 'Waspada', 2: 'Bahaya'}
safety_counts = df['safety_class'].map(safety_labels).value_counts()
colors = ['#22c55e', '#f59e0b', '#ef4444']
axes[0].pie(safety_counts.values, labels=safety_counts.index, colors=colors,
            autopct='%1.1f%%', startangle=90)
axes[0].set_title('Distribusi Kelas Keamanan')

# Gradient vs Safety
for cls in [0, 1, 2]:
    subset = df[df['safety_class'] == cls]
    axes[1].scatter(subset['gradient'], subset['effective_visibility'],
                    label=safety_labels[cls], alpha=0.6, s=30)
axes[1].set_title('Kemiringan vs Jarak Pandang')
axes[1].set_xlabel('Kemiringan (derajat)')
axes[1].set_ylabel('Jarak Pandang Efektif (km)')
axes[1].legend()

# Vehicle comparison
vehicle_safety = df.groupby(['vehicle_type', 'safety_class']).size().unstack(fill_value=0)
vehicle_safety.columns = [safety_labels[c] for c in vehicle_safety.columns]
vehicle_safety.plot(kind='bar', stacked=True, ax=axes[2], color=colors)
axes[2].set_title('Keamanan per Jenis Kendaraan')
axes[2].set_xlabel('Kendaraan')
axes[2].set_ylabel('Jumlah Sampel')
axes[2].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig('./figures/03_route_eda.png', dpi=150)
plt.show()
```

## 2. Train Model
