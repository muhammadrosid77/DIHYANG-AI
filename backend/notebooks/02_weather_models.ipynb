## Models in this notebook:
1. **Temperature Prediction** â€” Random Forest Regressor
2. **Rain Prediction** â€” Gradient Boosting Classifier  
3. **Tourism Risk Classification** â€” Gradient Boosting Classifier

```python
import sys
sys.path.insert(0, '..')

from app.models.train_weather_model import (
    load_historical_data, engineer_features,
    train_temperature_model, train_rain_classifier, train_risk_classifier,
    save_evaluation_report, FEATURE_COLS
)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

print('Modules loaded!')
```

## 1. Data Preparation & Feature Engineering

```python
# Load data
df = load_historical_data()

# Feature engineering
df = engineer_features(df)

print(f'\nFeature columns ({len(FEATURE_COLS)}):')
for i, col in enumerate(FEATURE_COLS, 1):
    print(f'  {i:2d}. {col}')
```

## 2. Model 1: Temperature Prediction (Random Forest Regressor)

```python
temp_model, temp_scaler, temp_metrics = train_temperature_model(df)

# Visualize feature importance
importances = pd.Series(temp_model.feature_importances_, index=FEATURE_COLS)
importances_sorted = importances.sort_values(ascending=True)

plt.figure(figsize=(10, 6))
importances_sorted.plot(kind='barh', color='#00b4cc')
plt.title('Feature Importance â€” Temperature Model')
plt.xlabel('Importance')
plt.tight_layout()
plt.savefig('./figures/02_temp_importance.png', dpi=150)
plt.show()
```

## 3. Model 2: Rain Prediction (Gradient Boosting Classifier)

```python
rain_model, rain_scaler, rain_metrics = train_rain_classifier(df)

# Visualize feature importance
importances = pd.Series(rain_model.feature_importances_, index=FEATURE_COLS)
importances_sorted = importances.sort_values(ascending=True)

plt.figure(figsize=(10, 6))
importances_sorted.plot(kind='barh', color='#3b82f6')
plt.title('Feature Importance â€” Rain Prediction Model')
plt.xlabel('Importance')
plt.tight_layout()
plt.savefig('./figures/02_rain_importance.png', dpi=150)
plt.show()
```

## 4. Model 3: Tourism Risk Classification

```python
risk_model, risk_scaler, risk_metrics = train_risk_classifier(df)
print('\nâœ… All weather models trained successfully!')
```

## 5. Save Models & Evaluation Report

```python
report = save_evaluation_report(temp_metrics, rain_metrics, risk_metrics)

print('\nðŸ“Š SUMMARY REPORT')
print('=' * 50)
print(f'Temperature RÂ²:  {temp_metrics["r2"]:.4f}')
print(f'Rain F1 Score:   {rain_metrics["f1"]:.4f}')
print(f'Risk Accuracy:   {risk_metrics["accuracy"]:.4f}')
```
