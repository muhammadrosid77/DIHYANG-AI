## Arsitektur NLP DITA

DITA menggunakan pendekatan **Hybrid NLP**:
1. **Gemini API** â€” Large Language Model untuk respons cerdas dan kontekstual

- Respons yang **akurat** (grounded pada data riset lapangan)

## 5. Danger Zone Analysis

```python
import matplotlib.pyplot as plt

zones = pd.DataFrame(DANGER_ZONES)

fig, ax = plt.subplots(figsize=(10, 5))
colors = ['#ef4444' if r == 'TINGGI' else '#f59e0b' for r in zones['risk']]
bars = ax.barh(zones['name'], zones['gradient_degree'], color=colors, edgecolor='white')
ax.set_xlabel('Kemiringan (derajat)')
ax.set_title('Zona Bahaya Wisata Dieng â€” Kemiringan Jalur')
ax.axvline(x=15, color='orange', linestyle='--', label='Batas Waspada (15Â°)')
ax.axvline(x=30, color='red', linestyle='--', label='Batas Bahaya (30Â°)')
ax.legend()

for bar, grad in zip(bars, zones['gradient_degree']):
    ax.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2,
            f'{grad}Â°', va='center', fontweight='bold')

plt.tight_layout()
plt.savefig('./figures/04_danger_zones.png', dpi=150)
plt.show()
```
