```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['font.size'] = 11

print('Libraries loaded successfully!')
```

## 1. Load Dataset

```python
with open('../app/data/dieng_historical_2023.json', 'r') as f:
    raw = json.load(f)

df = pd.DataFrame({
    'datetime': pd.to_datetime(raw['hourly']['time']),
    'temperature': raw['hourly']['temperature_2m'],
    'precipitation': raw['hourly']['precipitation']
})
df.set_index('datetime', inplace=True)

print(f'ðŸ“ Lokasi: Lat {raw["latitude"]}, Lon {raw["longitude"]}')
print(f'â›°ï¸  Elevasi: {raw["elevation"]}m')
print(f'ðŸ“… Periode: {df.index.min()} â€” {df.index.max()}')
print(f'ðŸ“Š Total records: {len(df)}')
print(f'\n{df.describe()}')
```

## 2. Distribusi Suhu

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histogram
axes[0].hist(df['temperature'], bins=50, color='#00b4cc', edgecolor='white', alpha=0.8)
axes[0].axvline(df['temperature'].mean(), color='red', linestyle='--', label=f'Mean: {df["temperature"].mean():.1f}Â°C')
axes[0].set_title('Distribusi Suhu Dieng 2023')
axes[0].set_xlabel('Suhu (Â°C)')
axes[0].set_ylabel('Frekuensi')
axes[0].legend()

# Boxplot per bulan
df['month'] = df.index.month
df.boxplot(column='temperature', by='month', ax=axes[1])
axes[1].set_title('Suhu per Bulan')
axes[1].set_xlabel('Bulan')
axes[1].set_ylabel('Suhu (Â°C)')
plt.suptitle('')

# Pola harian
df['hour'] = df.index.hour
hourly_mean = df.groupby('hour')['temperature'].mean()
axes[2].plot(hourly_mean.index, hourly_mean.values, 'o-', color='#f59e0b', linewidth=2)
axes[2].fill_between(hourly_mean.index, hourly_mean.values, alpha=0.2, color='#f59e0b')
axes[2].set_title('Pola Suhu Harian (Rata-rata)')
axes[2].set_xlabel('Jam')
axes[2].set_ylabel('Suhu (Â°C)')
axes[2].set_xticks(range(0, 24, 2))

plt.tight_layout()
plt.savefig('./figures/01_temperature_dist.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'\nðŸŒ¡ï¸ Suhu Min: {df["temperature"].min():.1f}Â°C')
print(f'ðŸŒ¡ï¸ Suhu Max: {df["temperature"].max():.1f}Â°C')
print(f'ðŸŒ¡ï¸ Suhu Mean: {df["temperature"].mean():.1f}Â°C')
print(f'ðŸŒ¡ï¸ Suhu Std: {df["temperature"].std():.1f}Â°C')
```

## 3. Analisis Presipitasi (Curah Hujan)

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Curah hujan bulanan
monthly_rain = df.groupby('month')['precipitation'].sum()
axes[0].bar(monthly_rain.index, monthly_rain.values, color='#3b82f6', edgecolor='white')
axes[0].set_title('Total Curah Hujan per Bulan (2023)')
axes[0].set_xlabel('Bulan')
axes[0].set_ylabel('Total Presipitasi (mm)')
axes[0].set_xticks(range(1, 13))

# Pola hujan per jam
hourly_rain = df.groupby('hour')['precipitation'].mean()
axes[1].bar(hourly_rain.index, hourly_rain.values, color='#6366f1', edgecolor='white')
axes[1].set_title('Rata-rata Presipitasi per Jam')
axes[1].set_xlabel('Jam')
axes[1].set_ylabel('Presipitasi (mm)')
axes[1].set_xticks(range(0, 24, 2))

plt.tight_layout()
plt.savefig('./figures/01_precipitation_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

rain_hours = len(df[df['precipitation'] > 0.1])
print(f'ðŸŒ§ï¸ Total jam hujan: {rain_hours} ({rain_hours/len(df)*100:.1f}%)')
print(f'ðŸŒ§ï¸ Total curah hujan: {df["precipitation"].sum():.1f} mm')
print(f'ðŸŒ§ï¸ Bulan paling basah: Bulan {monthly_rain.idxmax()} ({monthly_rain.max():.0f} mm)')
print(f'â˜€ï¸ Bulan paling kering: Bulan {monthly_rain.idxmin()} ({monthly_rain.min():.0f} mm)')
```

## 4. Identifikasi Zona Risiko Cuaca

Analisis kondisi cuaca yang berbahaya bagi wisatawan:

```python
# Identifikasi kondisi berbahaya
extreme_cold = df[df['temperature'] < 8]
heavy_rain = df[df['precipitation'] > 2.0]
combined_danger = df[(df['temperature'] < 8) & (df['precipitation'] > 0.5)]

print('âš ï¸ ANALISIS RISIKO CUACA DIENG 2023')
print('=' * 50)
print(f'ðŸ¥¶ Jam dengan suhu < 8Â°C (risiko hipotermia): {len(extreme_cold)} jam ({len(extreme_cold)/len(df)*100:.1f}%)')
print(f'ðŸŒ§ï¸ Jam dengan hujan > 2mm (hujan lebat): {len(heavy_rain)} jam ({len(heavy_rain)/len(df)*100:.1f}%)')
print(f'ðŸ”´ Jam dengan KOMBINASI (dingin + hujan): {len(combined_danger)} jam ({len(combined_danger)/len(df)*100:.1f}%)')

# Jam paling berbahaya
risk_by_hour = df.copy()
risk_by_hour['is_dangerous'] = (df['temperature'] < 10) | (df['precipitation'] > 1.0)
danger_hourly = risk_by_hour.groupby('hour')['is_dangerous'].mean() * 100

plt.figure(figsize=(12, 4))
colors = ['#ef4444' if v > 30 else '#f59e0b' if v > 15 else '#22c55e' for v in danger_hourly.values]
plt.bar(danger_hourly.index, danger_hourly.values, color=colors, edgecolor='white')
plt.title('Persentase Risiko Cuaca per Jam (Suhu < 10Â°C ATAU Hujan > 1mm)')
plt.xlabel('Jam')
plt.ylabel('% Waktu Berbahaya')
plt.xticks(range(0, 24))
plt.legend(['> 30% = ðŸ”´ Bahaya', '15-30% = âš ï¸ Waspada', '< 15% = âœ… Aman'])
plt.tight_layout()
plt.savefig('./figures/01_risk_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Korelasi dan Lag Analysis

```python
# Autocorrelation suhu
from pandas.plotting import autocorrelation_plot

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Lag plot
lag_corrs = [df['temperature'].autocorr(lag=i) for i in range(1, 49)]
axes[0].bar(range(1, 49), lag_corrs, color='#00b4cc')
axes[0].set_title('Autocorrelation Suhu (Lag 1-48 jam)')
axes[0].set_xlabel('Lag (jam)')
axes[0].set_ylabel('Korelasi')
axes[0].axhline(y=0, color='black', linestyle='-', linewidth=0.5)

# Scatter: suhu vs presipitasi
axes[1].scatter(df['temperature'], df['precipitation'], alpha=0.1, s=5, color='#6366f1')
axes[1].set_title('Korelasi Suhu vs Presipitasi')
axes[1].set_xlabel('Suhu (Â°C)')
axes[1].set_ylabel('Presipitasi (mm)')

plt.tight_layout()
plt.savefig('./figures/01_correlation.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Korelasi suhu-presipitasi: {df["temperature"].corr(df["precipitation"]):.3f}')
print(f'Autocorrelation lag-1h: {df["temperature"].autocorr(1):.3f}')
print(f'Autocorrelation lag-24h: {df["temperature"].autocorr(24):.3f}')
```
